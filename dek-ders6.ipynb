```r
# http://gekkoquant.com/2012/05/26/neural-networks-with-r-simple-example/

library("neuralnet")
```

```r
runif(10,min=0, max=20)
```

```r
girdi <-  data.frame(runif(50, min=0, max=100))
cikti <- sqrt(girdi)
```

```r
testveri <- cbind(girdi,cikti)
colnames(testveri) <- c("Girdiler","Ciktilar")
```

```r
head(testveri)
```

```r
ag.karekok <- neuralnet(Ciktilar~Girdiler,testveri, hidden=10, threshold=0.01)
```

```r
plot(ag.karekok)
```

```r
kontrolveri <- data.frame((1:10)^2) #Generate some squared numbers
ag.sonuc <- compute(ag.karekok, kontrolveri) #Run them through the neural network
```

```r
 
#Lets see what properties net.sqrt has
ls(ag.sonuc)
 
#Lets see the results
print(ag.sonuc$net.result)
```
