```r
dg<-read.xls("turkey-natgas-2010-2017.xls",  skip=3) 
```

```r
head(dg)
```

```r
dg<-ts(dg)
```

```r
plot(dg[,3])
```

```r
aylik_dg<-ts(dg[,3],start = c(2010,01), frequency = 12)
```

```r
head(aylik_dg)
```

```r
plot(aylik_dg)
```

```r
cycle(aylik_dg)
```

```r
plot(decompose(aylik_dg))
```

```r
plot(stl(aylik_dg,s.window = "periodic"))
```

```r
library(forecast)
```

```r
fit_dg<-ses(aylik_dg)
```

```r
plot(forecast(fit_dg, h=10))
```

```r
fit_dg<-ets(aylik_dg)
```

```r
plot(forecast(fit_dg, h=10))
```

```r
fit_dg<-hw(aylik_dg)
plot(forecast(fit_dg, h=10))
```

```r
fit_dg<-hw(aylik_dg,seasonal = "multiplicative")
plot(forecast(fit_dg, h=20))
```

```r
forecast(fit_dg, h=20)
```

```r
sonuc<-forecast(fit_dg, h=10)
```

```r
fit_dg<-hw(aylik_dg,seasonal = "multiplicative", h=100)
forecast(fit_dg, h=100)
```
